```python
import numpy as np
import pandas as pd
import pickle
import random
import matplotlib.pyplot as plt

import torch
import torch.autograd as autograd
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
with open('../../outputs/classifierX3_plotres.pkl', 'rb') as f:
    t0 = pickle.load(f)
```

```python
def plot_av_loss(t):
    plt.plot(t, color='red')
    plt.xlabel('epoch')
    plt.ylabel('average loss')
    plt.show()

def plot_acc(t):
    plt.plot(t, color='orange')
    plt.xlabel('epoch')
    plt.ylabel('validation accuracy')
    plt.show()

def plot_inst_loss(t):
    plt.plot(t, color='c')
    plt.xlabel('every other iteration')
    plt.ylabel('propotional to inst loss')
    plt.show()

def plot_all(tup):
    print('best accuracy is ', np.max(tup[2]))
    plt.figure()
    plot_inst_loss(tup[0])
    plt.figure()
    plot_av_loss(tup[1])
    plt.figure()
    plot_acc(tup[2])
```

```python
plot_all(t0)
```
